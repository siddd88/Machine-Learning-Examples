```python
import get_data as gd

from Logistic_Reg import LogisticRegressionGD

import matplotlib.pyplot as plt 

%matplotlib inline
```

```python
X_train_01_subset = gd.X_train[(gd.y_train == 0) | (gd.y_train == 1)]
y_train_01_subset = gd.y_train[(gd.y_train == 0) | (gd.y_train == 1)]

lrgd = LogisticRegressionGD(eta=0.05, n_iter=1000, random_state=1)
lrgd.fit(X_train_01_subset,
         y_train_01_subset)

gd.plot_decision_regions(X=X_train_01_subset, 
                      y=y_train_01_subset,
                      classifier=lrgd)

plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')

plt.tight_layout()
```
